- Carilah derivative dari sebuah fungsi:
  - $y = x^{2} + 2x + 1$
  - $y = 4x^{3} - 3x^{2} + 2x -1$

```python
#perhitungan menggunakan penurunan symbolic

import sympy as sy
x = sy.Symbol('x')
x
y = x**2 + 2*x + 1
y2 = 4*x**3 - 3*x**2 + 2*x -1 

dy = y.diff()
print ('hasil dari derivative fungsi y : ' ,dy)

dy2 = y2.diff()
print ('hasil dari derivative fungsi y2 : ',dy2)
```

```python
#perhitungan menggunakan penurunan numerik

import numpy as np

x= np.linspace(-10,10, 100)
y = x**2 + 2*x + 1
y2 = 4*x**3 - 3*x**2 + 2*x -1 


dy = np.gradient (y,x )
print ('hasil dari derivative fungsi y : ' ,dy)

dy2 = np.gradient (y2, x)
print ('hasil dari derivative fungsi y2 : ',dy2)
```

```python
#plot turunan fungsi 

import matplotlib.pyplot as plt
#fungsi awal
plt.plot(x,y)
```

```python
#plot gambar x terhadap turunan fungsi y
plt.plot(x, dy)
```

```python
#plot gambar x terhadap turunan fungsi y2
plt.plot(x, dy2)
```
