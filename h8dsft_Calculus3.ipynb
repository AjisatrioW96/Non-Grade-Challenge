- Carilah gradient dari sebuah fungsi:
  - $3x+4y=5$

```python
import sympy as sy
x,y= sy.symbols('x y')

f = 3*x + 4*y - 5
```

```python
#mencari turunan parsial dimasukkan kedalma array

from sympy.tensor.array import derive_by_array

grad = derive_by_array(f, (x, y))
grad
```

```python
#mencari turunan parsial numerik

import numpy as np
def f(x,y): 
  return 3*x + 4*y - 5


x=np.linspace(-50,50)
y=np.linspace(-50,50)

#Mengkombinasikan semua kemungkinan dan di pack ke dalam matrix
z=np.array( [[f(i,j) for i in x] for j in y] )

#Ada 2 output do x dan do y
dx,dy=np.gradient(z)
```

```python
#plot z, dx , dy  turunan fungsi parsial secara gambar dan plot

import matplotlib.pyplot as plt

fig, ax = plt.subplots(ncols=5,figsize=(17,5))
ax[0].imshow(z)
ax[0].set_title('f(x,y)')
ax[1].imshow(dx)
ax[1].set_title('dx')
ax[2].imshow(dy)
ax[2].set_title('dy')
ax[3].plot(dx, y)
ax[4].plot(x, dy)
```
